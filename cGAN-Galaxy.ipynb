```python
#import relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Dense,Flatten,Reshape,Lambda,LeakyReLU,multiply,Dropout,BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as K 
from tensorflow.keras import Model
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
print(tf.__version__)
#from tensorflow.python.keras.utils import to_categorical
from tensorflow.python.keras.callbacks import ModelCheckpoint
from tensorflow.python.keras.layers.merge import concatenate as concat
import seaborn as sns
```

```python
df_gal = pd.read_csv('MyTable_Galaxy_Dato1998.csv') #load the SDSS spectroscopic data
df_gal = df_gal.fillna('None')
```

```python
for col in ['mag_g','mag_r','mag_i','mag_z']: #clean the data
    df_gal=df_gal.where(df_gal[col] >-1000)
df_gal = df_gal.dropna()
```

```python
df_gal=df_gal.where(df_gal['cls'] == 'GALAXY') #choose galaxies only
df_gal = df_gal.dropna()
```

```python
features = df_gal[['mag_g','mag_r','mag_i','mag_z']].to_numpy()#select features
```

```python
targets = df_gal[['z','mass']].to_numpy() #select labels
```

```python
#train-validation-test split
x_train = features[:round(len(features)*0.8)]
y_train = targets[:round(len(targets)*0.8)]
x_valid = features[round(len(features)*0.8):-1000]
y_valid = targets[round(len(targets)*0.8):-1000]
x_test = features[-1000:]
y_test = targets[-1000:]
```

```python
X_train = np.copy(x_train)
X_valid = np.copy(x_valid)
```

```python
codings_size = 100  #choose the size of encodings
num_classes = 10 #number of labels
```

```python
#get the shapes of features and targets
n_x = x_train.shape[1]
n_y = y_train.shape[1]

X = Input(shape=[codings_size,])
label = Input(shape=[n_y,])
pixels = Input(shape=[n_x,])
```

```python
#Create the generator
inputs = concat([X,label])
z= Dense(200)(inputs)
z = LeakyReLU()(z)
z=Dropout(0.2)(z)
z= Dense(150)(z)
z = LeakyReLU()(z)
z=Dropout(0.2)(z)
z= Dense(100)(z)
z = LeakyReLU()(z)
z=Dropout(0.2)(z)
z = Dense(50)(z)
z = LeakyReLU()(z)
z=Dropout(0.2)(z)
outputs = Dense(n_x)(z)
generator = Model(inputs = [X,label], outputs = [outputs])
```

```python
#Create the discriminator
inputs = concat([pixels,label])
z= Dense(250)(inputs)
z = LeakyReLU()(z)
z=Dropout(0.2)(z)
z = Dense(200)(z)
z = LeakyReLU()(z)
z = Dense(150)(z)
z=Dropout(0.2)(z)
z = LeakyReLU()(z)
z=Dropout(0.2)(z)
z = Dense(100)(z)
z = LeakyReLU()(z)
z=Dropout(0.2)(z)
valids = Dense(1, activation = 'sigmoid')(z)
disc = Model(inputs = [pixels,label], outputs = [valids])
```

```python
#link uo the cGAN
img = generator([X,label])
valids = disc([img,label])
gan = Model(inputs=[X,label], outputs = [valids])
```

```python
disc.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop') #loss of the discriminator
disc.trainable = False
gan.compile(loss='mean_squared_error', optimizer='rmsprop') #loss of the generator
```

```python
batch_size = 32#define a batch
#create a tf.datasets using the traing and testing data and shuffle
x_dataset = tf.data.Dataset.from_tensor_slices(x_train)
x_dataset = x_dataset.batch(batch_size,drop_remainder=True).prefetch(1)

y_dataset = tf.data.Dataset.from_tensor_slices(y_train)
y_dataset = y_dataset.batch(batch_size,drop_remainder=True).prefetch(1)
```

```python
#A function to train the cGAN
def train_gan(gan, dataset,y_dataset, batch_size, codings_size, n_epochs=100):
    g, d = gan.layers[2],  gan.layers[3]
    for epoch in range(n_epochs):
        for (batch,y_batch) in zip(dataset,y_dataset):
            #step 1: train the discriminatoe
            noise = tf.random.normal(shape = [batch_size, codings_size])#create random noise
            generated_images = g([noise,y_batch])#feed it to the generator
            batch=tf.cast(batch, dtype=tf.float32)
            y_batch=tf.cast(y_batch, dtype=tf.float32)
            X_fake_and_real = tf.concat([generated_images, batch], axis = 0)#concatenate real data with the generated data
            y1 = tf.constant([[0.]]*batch_size+[[1.]]*batch_size)#make the labels
            d.trainable = True
            Y_batch = tf.concat([y_batch, y_batch], axis = 0)
            d.train_on_batch([X_fake_and_real,Y_batch], y1)#train the discriminator
            
            #step 2: train the generator
            noise = tf.random.normal(shape = [batch_size, codings_size])#create random noise
            y2 = tf.constant([[1.]] * batch_size)#set labels as real
            d.trainable = False
            gan.train_on_batch([noise,y_batch],y2)#train the generator
        print(epoch)
        if epoch%1o == 0: #after every 10th epoch display the generated magnitude distribution
            print('Epoch: '+str(epoch+1)+':')
            n = 1000
            labels = y_test
            new_codings = np.random.normal(size=[n,codings_size])
            new_codings = tf.convert_to_tensor(new_codings, dtype=tf.float32)
            labels = tf.convert_to_tensor(labels, dtype=tf.float32)
            galaxies = generator([new_codings,labels]).numpy()
            diff = x_test-galaxies
            means = np.mean(diff,axis=0)
            print(means)
            for col in range(4):
                plt.hist(np.transpose(galaxies)[col], alpha=0.5,bins=50)
                plt.xlabel('Magnitude')
            plt.show()
```
